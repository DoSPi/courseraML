```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import expon
from scipy.stats import norm
from math import sqrt
%matplotlib inline
```

Генерируем выборку размера 1000 из экспоненциального распределения с параметром масштаба 1

```python
sample = expon.rvs(size=1000)
```

Строим гистограмму выборки  и теоретическую плотность на графике

```python
plt.hist(sample,density = True, bins = 30)
x = np.linspace(0,10,100)
plt.plot(x, expon.pdf(x), color='red')
plt.xlabel('X')
plt.ylabel('плотность')
```

т. к. X ~ Exp(1), то E(X) = 1; D(X) = 1;
Возьмем  по 1000 выборок объема  n = 5, 10 , 50 и сравним распределение их выборочных средних с нормальным распределением N(1, 1/n)

```python
n = 5
x = []
for i in range(0, 1000):
    x.append(np.sum(expon.rvs(size = n))/ n)
plt.hist(x,density = True, bins = 20)
plt.xlabel('X')
plt.ylabel('плотность')
plt.title('N = 5')
x = np.linspace(0,3,100)
plt.plot(x,norm.pdf(x,1, sqrt(1/n)), color = 'red')
```

```python
n = 10
x = []
for i in range(0, 1000):
    x.append(np.sum(expon.rvs(size = n))/ n)
plt.hist(x,density = True, bins = 20)
plt.xlabel('X')
plt.ylabel('плотность')
plt.title('N = 10')
x = np.linspace(0,3,100)
plt.plot(x,norm.pdf(x,1, sqrt(1/n)), color = 'red')
```

```python
n = 50
x = []
for i in range(0, 1000):
    x.append(np.sum(expon.rvs(size = n))/ n)
plt.hist(x,density = True, bins = 20)
plt.xlabel('X')
plt.ylabel('плотность')
plt.title('N = 50')
x = np.linspace(0,3,100)
plt.plot(x,norm.pdf(x,1, sqrt(1/n)), color = 'red')
```

При увеличении размера выборки распределение выборочных средних стремится к нормальному N(1, 1/n), что
соответствует ЦПТ
